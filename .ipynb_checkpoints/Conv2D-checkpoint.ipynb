```python
%matplotlib inline

img_path="bf.jpg"
img=Image.open(img_path)#.convert("L")
plt.imshow(img)
to_tensor=transforms.ToTensor()
to_img=transforms.ToPILImage()

#img.show()
```

```python
img=to_tensor(img.convert("L"))
img=torch.unsqueeze(img,0)
c=nn.Conv2d(1,1,kernel_size=3,padding=1)
c.weight=torch.nn.Parameter(torch.Tensor(
    [[[
        [-1,0,1],
        [-2,0,2],
        [-1,0,1]]]] #sobel operator

))
img_conv2d=c(img)
img_conv2d=torch.squeeze(img_conv2d,0)
img_conv2d=to_img(img_conv2d)
plt.imshow(img_conv2d)
#img_conv2d.show()
```

```python
from torchvision.models import resnet18
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

plt.ion()
totensor=transforms.ToTensor()
toimg=transforms.ToPILImage()
net=resnet18(pretrained=True)
img=Image.open("bf.jpg")
img_tensor=totensor(img).unsqueeze(0)
plt.imshow(img)
           #img.show()
#plt.clf()
```

```python
f1=net.conv1(img_tensor)
plt.figure(figsize=(10,10))
for p in range(4):
    f1_img_tensor = f1[0,p,:,:]
    f1_img=toimg(f1_img_tensor)
    plt.subplot(220+p+1)
    plt.imshow(f1_img)
    #plt.pause(1)
plt.show()
#plt.clf()
```

```python
f2=net.bn1(f1)
plt.figure(figsize=(10,10))
for p in range(4):
    f2_img_tensor=f2[0,p,:,:]
    f2_img=toimg(f2_img_tensor)
    plt.subplot(220+p+1)
    plt.imshow(f2_img)
    #plt.pause(1)
plt.show()
#plt.clf()
```

```python
f3=net.relu(f2)
plt.figure(figsize=(10,10))
for p in range(4):
    f3_img_tensor=f3[0,p,:,:]
    f3_img=toimg(f3_img_tensor)
    plt.subplot(220+p+1)
    plt.imshow(f3_img)
    #plt.pause(1)
plt.show()
#plt.clf()
```

```python
f4=net.maxpool(f3)
plt.figure(figsize=(10,10))
for p in range(4):
    f4_img_tensor=f4[0,p,:,:]
    f4_img=toimg(f4_img_tensor)
    plt.subplot(220+p+1)
    plt.imshow(f4_img)
    #plt.pause(1)
plt.show()
plt.pause(10)
#plt.ioff()
```
